Task 2: 

Are there any null values in the Grc_df? If yes, then in which columns and how many? Finally, handle these null values using any strategy shown during the labs.

```python
# Sjekker for null verdier i hver kolonne
null_counts = Grc_df.isnull().sum()

print("Number of null values in each column:")
print(null_counts)
```

Svar: Ja, det er null verdier i Grc_df DataFrame. De er blitt funnet i f√∏lgene kolonner:

Item_Weight: har 1463 null verdier,
Outlet_Size: har 2410 null verdier

```python
# Strategien vi bruker her er √• flatfylle verdiene med gjennomsnittet
# H√•ndterer null verdiene for 'Item_Weight' ved √• fylle inn gjennomsnittet
mean_Weight = Grc_df['Item_Weight'].mean()
Grc_df['Item_Weight'].fillna(mean_Weight, inplace=True)

# H√•ndterer null verdier for 'Outlet_Size' ved √• fyllle ut med den som er mest brukt
mode_size = Grc_df['Outlet_Size'].mode()[0]
Grc_df['Outlet_Size'].fillna(mode_size, inplace=True)

# Verifiserer om null verdiene har blitt behandlet
print("\nNumber of null values after handling:")
print(Grc_df.isnull().sum())
```

Using Seaborn (‚Äúggplot style‚Äù) create a (2,1) subplot of a box plot showing 5-point summary of the column Item_MRP for SupType_1 and SupType_2. 
Which Outlet Type has a higher median MRP? Are there any outliers?

Task 8:

Concatenate the DataFrames SupType_1 and SupType_2 to create a new DataFrame Grc_Concat_df and sort it in ascending order based on the column Item_Outlet_Sales.
What is the Outlet_Location_type of the store with the lowest sales?

```python
# Concatenate the DataFrames SupType_1 and SupType_2 to create a 
# new DataFrame Grc_Concat_df and sort it in ascending order based on the column Item_Outlet_Sales.

# Sl√•r sammen SupType_1 og SupType_2 til en ny DataFrame
# ignore_index=True nullstiller indeksen fra 0 til n-1 i den nye DataFrame-en
Grc_Concat_df = pd.concat([SupType_1, SupType_2], ignore_index=True)

print("The DataFrames are now merged.")
print(f"The new DataFrame Grc_Concat_df has {Grc_Concat_df.shape[0]} rows.")

# Sorterer DataFrame-en i stigende rekkef√∏lge basert p√• 'Item_Outlet_Sales'
# inplace=True endrer DataFrame-en direkte uten √• lage en ny kopi
Grc_Concat_df.sort_values(by='Item_Outlet_Sales', inplace=True)
print("Grc_Concat_df has been sorted by Item_Outlet_Sales.")
```

```python
# What is the Outlet_Location_type of the store with the lowest sales?

# Henter ut den f√∏rste raden (som n√• har det laveste salget)
row_with_lovest_sale = Grc_Concat_df.iloc[0]

# Henter ut verdien fra 'Outlet_Location_Type'-kolonnen i den raden

# Henter ut verdien fra 'Outlet_Location_Type'-kolonnen i den raden
location_type_lowest_sale = row_with_lovest_sale['Outlet_Location_Type']

# For √• verifisere, ser vi p√• de 5 √∏verste radene i den sorterte DataFrame-en
print("\nTop 5 rows in the sorted DataFrame (shows lowest sales):")
display(Grc_Concat_df.head())

print(f"\nQuestion: What is the Outlet_Location_type of the store with the lowest sales?")
print(f"Answer: {location_type_lowest_sale}")
```

Task 10:

Cut the column Item_Weight of Grc_Concat_df into 10 buckets and compute each bucket's mean, minimum, maximum, and count.

```python
# Trinn 1: Oppretter de 10 b√∏ttene fra kolonnen 'Item_Weight'.
# pd.cut returnerer en serie med kategoriserte data som vi kan bruke til gruppering.
weight_buckets = pd.cut(Grc_Concat_df['Item_Weight'], bins=10)

# Trinn 2: Grupperer den opprinnelige DataFrame etter disse nye b√∏ttene.
# Velger deretter kolonnen 'Item_Weight' fra disse gruppene og utf√∏rer aggregeringene.
bucket_analysis = Grc_Concat_df.groupby(weight_buckets, observed=True)['Item_Weight'].agg(['mean','min','max','count'])

# Vsier resultatet for verifisering
print("Analysis of Item_Weight Buckets:")
display(bucket_analysis)
```
