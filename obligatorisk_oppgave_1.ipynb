Task 2: 

Are there any null values in the Grc_df? If yes, then in which columns and how many? Finally, handle these null values using any strategy shown during the labs.

```python
# Sjekker for null verdier i hver kolonne
null_counts = Grc_df.isnull().sum()

print("Number of null values in each column:")
print(null_counts)
```

Svar: Ja, det er null verdier i Grc_df DataFrame. De er blitt funnet i følgene kolonner:

Item_Weight: har 1463 null verdier,
Outlet_Size: har 2410 null verdier

```python
# Strategien vi bruker her er å flatfylle verdiene med gjennomsnittet
# Håndterer null verdiene for 'Item_Weight' ved å fylle inn gjennomsnittet
mean_Weight = Grc_df['Item_Weight'].mean()
Grc_df['Item_Weight'].fillna(mean_Weight, inplace=True)

# Håndterer null verdier for 'Outlet_Size' ved å fyllle ut med den som er mest brukt
mode_size = Grc_df['Outlet_Size'].mode()[0]
Grc_df['Outlet_Size'].fillna(mode_size, inplace=True)

# Verifiserer om null verdiene har blitt behandlet
print("\nNumber of null values after handling:")
print(Grc_df.isnull().sum())
```

Task 3:

How many unique Outlet Sizes are there in the Grc_df? 
Which outlet size is maximum, and which is minimum?

```python
# Deloppgave 3, del 1/2:
# How many unique Outlet Sizes are there in the Grc_df? 

# Hent antallene for hver unike verdi i kolonnen 'Outlet_Size'
outlet_size_counts = Grc_df['Outlet_Size'].value_counts()
# print(f"{outlet_size_counts}") # debug print
num_unique_sizes = len(outlet_size_counts)

print(f"There are {num_unique_sizes} unique Outlet Sizes in the DataFrame.")
print("\nHere are the counts for each size:")
print(outlet_size_counts)

```

```python
# Deloppgave 3, del 2/2:
# How many unique Outlet Sizes are there in the Grc_df? 

# Maksimum er den første indeksen i value_counts-serien
max_size = outlet_size_counts.index[0]

# Minimumet er den siste indeksen i value_counts-serien
min_size = outlet_size_counts.index[-1]

print(f"\nThe outlet size with the maximum number of outlets is: '{max_size}'")
print(f"The outlet size with the minimum number of outlets is: '{min_size}'")
```

Task 4: 

How many unique Item Fat Content types are in the Grc_df? List them. Do you see any issues with the Item Fat Content types? 
If yes, then handle this issue.

```python
# Oppgave 4: del 1/2 
# How many unique Item Fat Content types are in the Grc_df?

# Henter de unike verdiene og antallene deres for 'Item_Fat_Content'
fat_content_counts = Grc_df['Item_Fat_Content'].value_counts()

print(f"There are {len(fat_content_counts)} unique 'Item Fat Content' types.")
print("\nHere is a list of the types and their counts:")
print(fat_content_counts)
```

```python
# Oppgave 4: del 2/2
# Do you see any issues with the Item Fat Content types? 

# Steg 1: Definerer en "ordbok" for erstatningene
replacements = {
    'low fat': 'Low Fat',
    'LF': 'Low Fat',
    'reg': 'Regular'
}

# Steg 2: Bruker .replace() for å utføre endringer i kolonnen
Grc_df['Item_Fat_Content'].replace(replacements, inplace=True)


# Steg 3: Verifiser at problemet er løst ved å kjøre value_counts() på nytt
print("\n'Item Fat Content'-typer etter opprydding:")
print(Grc_df['Item_Fat_Content'].value_counts())
```
